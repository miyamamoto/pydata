```python
#世田谷区におけるAEDの場所を地図上にマッピングする。

import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

#　事前にAEDの場所が記録されたデータをダウンロードする。
# ダウンロード元は以下のURL
# http://www.city.shinagawa.tokyo.jp/ct/other000061100/366.csv

df = pd.read_csv("./aed/366.csv",encoding="SHIFT-JIS")
df.head()
```

```python
# pandas　DataFrameのデータ形式をgeopandas GeoDataFrame形式に変換する。

geometry = [Point(xy) for xy in zip( df[u"経度"],df[u"緯度"])]
gdf = gpd.GeoDataFrame(df, geometry=geometry,crs = {'init' :'epsg:4326'})
gdf.head()
```

```python
# for
import folium  

m = folium.Map([35.609,139.725], zoom_start=14,tiles='cartodbpositron')
folium.GeoJson(gdf).add_to(m)
m
```
